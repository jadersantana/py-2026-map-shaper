```python
import geopandas as gpd
from pathlib import Path

DATA_DIR = Path("../data").resolve()

shp_path = DATA_DIR / "input/ibge/uf/BR_UF_2024.shp"
output_path = DATA_DIR / "output/br_uf_2024_simplified.geojson"

gdf = gpd.read_file(shp_path)

# Simplificar geometria
gdf["geometry"] = gdf["geometry"].simplify(
    tolerance=0.01,   # você pode ajustar
    preserve_topology=True
)

gdf.to_file(output_path, driver="GeoJSON")

print("GeoJSON simplificado salvo.")
```

```python
import geopandas as gpd
import unicodedata
from pathlib import Path

def normalize(text):
    text = unicodedata.normalize("NFKD", text)
    text = text.encode("ASCII", "ignore").decode("utf-8")
    return text.upper()

DATA_DIR = Path("../data").resolve()

shp_path = DATA_DIR / "input/ibge/cidade/pr/PR_Municipios_2024.shp"
output_path = DATA_DIR / "output/PR_Municipios_2024_simplified_normalized.geojson"

gdf = gpd.read_file(shp_path)

# Normalizar nome do estado
gdf["NM_MUN"] = gdf["NM_MUN"].apply(normalize)

gdf["geometry"] = gdf["geometry"].simplify(
    tolerance=0.01,   # você pode ajustar
    preserve_topology=True
)

gdf.to_file(output_path, driver="GeoJSON")

print("GeoJSON normalizado salvo.")
```
